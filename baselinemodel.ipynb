```python
dataset = pd.concat([d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12,d13,d14,d15],ignore_index=True)
```

```python
dataset.describe()
```

```python
x_data = dataset.drop(['marker'],axis = 1 )
```

```python
y_data = dataset[['marker']]
```

```python
y_data = y_data.astype('category')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x_data,y_data,test_size = 0.2, random_state = 2)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators=100, random_state=0)
```

```python
x_train.dropna()
```

```python
x_train = x_train.reset_index()
```

```python
y_train = y_train.reset_index()
```

```python
x_train.drop(['index'],axis = 1 , inplace = True)
```

```python
x_train
```

# you need to remove infinite value and null values from the dataset, now what are infinite value you need to work on that , the error was that too big for float64

```python
x_train.fillna(999, inplace=True)
```

```python
y_train = y_train['marker']
```

```python
ada.fit(x_train,y_train)
```

```python
x_test.replace([np.inf, -np.inf], np.nan, inplace=True)
x_test.fillna(999, inplace=True)
```

```python
pred_ada = ada.predict(x_test)
```

```python
 from sklearn.metrics import classification_report

cr_ada = classification_report(y_test,pred_ada)
```
